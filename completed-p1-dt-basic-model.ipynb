## Partition Data into X and y

```python
y = data['Approved']
X = data.drop('Approved', axis = 1)
```

```python
# Check y
y.head(10)
```

```python
# Check X
X.head(10)
```

## Decision Tree

### 1/ Import Library

```python
# Classification
from sklearn.tree import DecisionTreeClassifier
```

```python
DecisionTreeClassifier
```

### 2/ Initiate an Instance

```python
dt_model = DecisionTreeClassifier(random_state = 42)
```

### 3/ Fit Model

```python
dt_model.fit(X,y)
```

### Pre Process Data
- Only Dummy Variable Creation 

```python
X = pd.get_dummies(X)
X.head()
```

```python
X.info()
```

### 3/ Fit Model (Again)

```python
dt_model.fit(X,y)
```

```python
dt_model.get_depth()
```

### 4/ Make Predictions

```python
y_predicted = dt_model.predict(X)
y_predicted
```

### 5/ Check Performance

#### Method 1
- score
- inputs : X and y

```python
dt_model.score(X, y)
```

#### Method 2
- Import from Metrics
- Call an Instance 
- Use Actual vs Predicted y to get Performacne

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y, y_predicted))
```
